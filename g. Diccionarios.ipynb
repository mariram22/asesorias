Es una colección de datos mutable que organiza sus elementos de acuerdo a una llave. 

Llave: un elemento ÚNICO que identifica un valor asociado a esta. 

Cada elemento de un diccionario contiene un par llave:'valor asociado'. Cada elemento separado por comas y encerrados entre llaves.

```python
p1 = {'Nombre':'Santiago','Edad':26,'Altura':1.86}
print(p1)
print(type(p1))
p2 = {'Nombre':'Alejandra','Edad':31,'Altura':1.75}
print(p2)
print(type(p2))
```

Agreguemos una nueva llave:

```python
p1['EPS']='Sura' # diccionario['llave']=valor
print(p1)
```

Modifiquemos ahora uno de sus valores:

```python
print(p1['EPS'])
p1['EPS']='Salud total'
print(p1)
```

Pasemos ahora a eliminar una de las llaves:

```python
p1.pop('EPS')
print(p1)
```

-----
Aclaración pop:

En realidad, esta herramienta no 'destruye' el elemento si no que lo 'saca', esto quiere decir que al final este elemento puede ser reasignado a otra variable.

Expliquemos esto con listas:

```python
lista = [i+10 for i in range(11)]
print(lista)
x=lista.pop(1)
print(lista)
print(x)
```

Para determinar la cantidad de llaves de un diccionario se hace lo siguiente

```python
cantidad = len(p1)
print(cantidad)
```

Para obtener los valores o las llaves de un diccionaro como listas se puede hacer lo siguiente:

```python
valores=list(p1.values())
print(valores)
llaves=list(p1.keys())
print(llaves)
```
