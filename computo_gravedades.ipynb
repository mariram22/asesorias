```python
import numpy as np
import math
import matplotlib.pyplot as plt
print(f"Numpy version: {np.__version__}")
```

```python
ph = (20 * 4 * (math.cos(math.pi/4)))
print (ph)
pv = (20 * (math.sin(math.pi/2))*4 - 1/2*9.81*4**2)
print (pv)
```

```python
tiempo = np.linspace(0, 4, num=100)
print(tiempo)
type(tiempo)
```

```python
X = 20 * tiempo * (math.cos(math.pi/4))
print(X)
```

```python
Y = 20 * (math.sin(math.pi/4))*tiempo - 1/2*9.81*tiempo**2
print(Y)
```

```python
plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Trayectoria del proyectil')
plt.show()
```
