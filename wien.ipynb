```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example: Given distribution (students are provided this array)
wavelengths = np.linspace(1e-7, 3e-6, 100)  # Wavelengths in meters
# Simulate spectral radiance for a blackbody (e.g., T = 5000K)
def planck_law_simple(wavelength, T):
    """Simplified version of Planck's Law for demonstration purposes."""
    return (2 * 6.626e-34 * 3e8**2) / (wavelength**5 * (np.exp(6.626e-34 * 3e8 / (wavelength * 1.38e-23 * T)) - 1))

T_true = 5000  # True temperature of the blackbody in Kelvin
spectral_radiance = planck_law_simple(wavelengths, T_true)

# Plot the spectral radiance
plt.plot(wavelengths * 1e9, spectral_radiance, label=f'T = {T_true}K')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Spectral Radiance')
plt.title('Blackbody Radiation Spectrum')
plt.grid(True)
plt.show()
```

```python
peak_index = np.argmax(spectral_radiance)
peak_wavelength = wavelengths[peak_index]

# Wien's displacement law: λ_max * T = 2.898e-3 m·K
wien_constant = 2.898e-3  # m·K
estimated_temperature = wien_constant / peak_wavelength

print(f"Peak wavelength: {peak_wavelength * 1e9:.2f} nm")
print(f"Estimated temperature: {estimated_temperature:.2f} K")
```
