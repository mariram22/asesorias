# <font color='red'> <center>PASO POR VALOR-REFERENCIA</center></font>
- *Paso por valor*: Se crea una copia local de la variable dentro de la función.
- *Paso por referencia*: Se maneja directamente la variable, los cambios realizados en la función le afectarán también fuera.

Tradicionalmente, **los tipos simples se pasan automáticamente por valor y los compuestos por referencia**.
- *Simples*: Enteros, flotantes, string, lógicos...
- *Compuestos*: Listas, diccionarios, tuplas...

##### Ejemplo paso por valor:

```python
def duplicar_valor(numero):
    numero*=2
n = 10
print(n)
duplicar_valor(n)
print(n)
```

##### Ejemplo paso por referencia:

```python
def duplicar_valores(numeros):
    for i in range(len(numeros)):
        numeros[i] *=2
        
ns = [10,50,100]
print(ns)
duplicar_valores(ns)
print(ns)
```

### Trucos:
Para modificar los tipos simples podemos devolverlos modificados y reasignarlos:

```python
def duplicar_valor(numero):
    return numero*2
n=30
print(n)
n = duplicar_valor(n)
print(n)
```

Y en el caso de los tipos compuestos, podemos evitar la modificación enviando una copia:

```python
def duplicar_valores(numeros):
    for i in range(len(numeros)):
        numeros[i] *= 2
ns = [10,50,100]
print(ns)
duplicar_valores(ns.copy()) 
print(ns)
```
