```python
import sympy as sy
import scipy as sp
import numpy as np
```

```python
x, a = sy.symbols('x a', real=True, positive=True)
```

```python
psi_0 = sy.exp((-a*x**2)/2) * sy.hermite(0,sy.sqrt(a) * x) * sy.sqrt(sy.sqrt(a)/(sy.sqrt(sy.pi)))
psi_1 = sy.exp((-a*x**2)/2) * sy.hermite(1,sy.sqrt(a) * x) * sy.sqrt(sy.sqrt(a)/(sy.sqrt(sy.pi)*2))
psi_2 = sy.exp((-a*x**2)/2) * sy.hermite(2,sy.sqrt(a) * x) * sy.sqrt(sy.sqrt(a)/(sy.sqrt(sy.pi)*8))
psi_3 = sy.exp((-a*x**2)/2) * sy.hermite(3,sy.sqrt(a) * x) * sy.sqrt(sy.sqrt(a)/(sy.sqrt(sy.pi)*48))
psi_4 = sy.exp((-a*x**2)/2) * sy.hermite(4,sy.sqrt(a) * x) * sy.sqrt(sy.sqrt(a)/(sy.sqrt(sy.pi)*384))
```

```python
sy.integrate(psi_0**2, (x, -sy.sqrt(1/a), sy.sqrt(1/a))).evalf()
```

```python
def probabilidad_prohibida(psi, n):
    region_prohibida = sy.sqrt(2 * (n + 1/2)/a)
    probabilidad = sy.integrate(psi**2, (x, region_prohibida, sy.oo))
    return 2 * probabilidad.evalf()
```

```python
psi_0 / sy.sqrt(a)
```

```python
sy.integrate(psi_0**2, (x, -sy.oo, sy.oo))
```

```python
print(probabilidad_prohibida(psi_0, 0))
print(probabilidad_prohibida(psi_1, 1))
print(probabilidad_prohibida(psi_2, 2))
print(probabilidad_prohibida(psi_3, 3))
print(probabilidad_prohibida(psi_4, 4))
```

```python
# i also want to plot the parabola of the potential in the same sy.plot
sy.plot((psi_0 ** 2).subs(a, 1), 0.25 * x**2, (x, -2, 2), ylabel='psi_0')
sy.plot((psi_1 ** 2).subs(a, 1), 0.25 * x**2, (x, -4, 4), ylabel='psi_1')
sy.plot((psi_2 ** 2).subs(a, 1), 0.25 * x**2, (x, -4, 4), ylabel='psi_2')
sy.plot((psi_3 ** 2).subs(a, 1), 0.25 * x**2, (x, -4, 4), ylabel='psi_3')
sy.plot((psi_4 ** 2).subs(a, 1), 0.25 * x**2, (x, -4, 4), ylabel='psi_4')
```
