```python
import sympy as sp
sp.init_printing()
```

```python
H = sp.Matrix([[2, sp.I, 0], [-sp.I, 1, 1], [0, 1, 0]])
H
```

```python
psi = sp.Matrix([sp.I, 7*sp.I, -2])
H * psi
```

```python
H.eigenvals()
```

```python
eigenvecs = H.eigenvects()
eigenvecs[0][2][0]
```
