```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn  # Función de Bessel del primer tipo

# Configurar la cuadrícula para coordenadas polares
n_r = 100  # Número de puntos en la dirección radial
n_theta = 100  # Número de puntos en la dirección angular
r = np.linspace(0, 10, n_r)  # Puntos radiales
theta = np.linspace(0, 2 * np.pi, n_theta)  # Puntos angulares de 0 a 2*pi

# Crear meshgrid para r y theta
R, Theta = np.meshgrid(r, theta)

# Ejemplo de solución con la función de Bessel: R_n(r) * Theta_m(theta)
n = 1  # Orden de la función de Bessel
m = 2  # Número de modos angulares
A = 1  # Amplitud

# Calcular la solución radial usando la función de Bessel del primer tipo
R_n = jn(n, R)

# Calcular la solución angular (una simple onda seno o coseno)
Theta_m = np.cos(m * Theta)

# Combinar las partes radial y angular para obtener la solución completa Phi(r, theta)
Phi = A * R_n * Theta_m

# Convertir a coordenadas cartesianas para graficar en 3D
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

# Crear una gráfica 3D de la solución
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
ax.plot_surface(X, Y, Phi, cmap='viridis')

# Etiquetar los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Potencial (Φ)')

plt.title(f'Solución de la Ecuación de Laplace Usando la Función de Bessel: n={n}, m={m}')
plt.show()
```
