```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn  # Funcin de Bessel del primer tipo

# Configurar la cuadrcula para coordenadas polares
n_r = 100  # Nmero de puntos en la direccin radial
n_theta = 100  # Nmero de puntos en la direccin angular
r = np.linspace(0, 10, n_r)  # Puntos radiales
theta = np.linspace(0, 2 * np.pi, n_theta)  # Puntos angulares de 0 a 2*pi

# Crear meshgrid para r y theta
R, Theta = np.meshgrid(r, theta)

# Ejemplo de solucin con la funcin de Bessel: R_n(r) * Theta_m(theta)
n = 1  # Orden de la funcin de Bessel
m = 2  # Nmero de modos angulares
A = 1  # Amplitud

# Calcular la solucin radial usando la funcin de Bessel del primer tipo
R_n = jn(n, R)

# Calcular la solucin angular (una simple onda seno o coseno)
Theta_m = np.cos(m * Theta)

# Combinar las partes radial y angular para obtener la solucin completa Phi(r, theta)
Phi = A * R_n * Theta_m

# Convertir a coordenadas cartesianas para graficar en 3D
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

# Crear una grfica 3D de la solucin
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
ax.plot_surface(X, Y, Phi, cmap='viridis')

# Etiquetar los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Potencial ()')

plt.title(f'Solucin de la Ecuacin de Laplace Usando la Funcin de Bessel: n={n}, m={m}')
plt.show()
```
